### Running actual model

```python
def load_data(file_path):
    df = pd.read_csv(file_path)
    return df

def sentiment_analysis(comments):
    device = 0 if torch.cuda.is_available() else -1

    sentiment_pipeline = pipeline("sentiment-analysis", device=device)

    results = sentiment_pipeline(comments)
    return results

def process_dataframe(file_path, comment_column, output_file):
    df = load_data(file_path)

    sentiments = sentiment_analysis(df[comment_column].tolist())

    df['Sentiment'] = [result['label'] for result in sentiments]
    df['Sentiment_Score'] = [result['score'] for result in sentiments]

    df.to_csv(output_file, index=False)

if __name__ == "__main__":
    input_file = '/Users/aa085/Music/Datastate.csv'
    output_file = '/Users/aa085/Music/Datastate_with_sentiments.csv'
    comment_column = 'Comments'
    process_dataframe(input_file, comment_column, output_file)
```
